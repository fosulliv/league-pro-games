```python
import numpy as np
import pandas as pd
import os
import requests
import bs4
import json
```

```python
url = 'https://gol.gg/game/stats/{}/page-game/'
first = 27390
last = 31280
htt = requests.get(url.format(first)).content
soup = bs4.BeautifulSoup(htt)
champs = soup.find_all('img', attrs={'class': 'champion_icon_medium rounded-circle'})
blue = champs[5:10]
red = champs[15:20]
blue = [i['alt'] for i in blue]
red = [i['alt'] for i in red]
blue_bans = champs[:5]
blue_bans = [i['alt'] for i in blue_bans]
red_bans = champs[10:15]
red_bans = [i['alt'] for i in red_bans]
red_bans
```

```python
games = {'blue_top': blue[0], 'blue_jg': blue[1], 'blue_mid': blue[2], 'blue_adc': blue[3],
         'blue_sup': blue[4], 'red_top': red[0], 'red_jg': red[1], 'red_mid': red[2], 'red_adc': red[3],
         'red_sup': red[4], 'blue_ban1': blue_bans[0], 'blue_ban2': blue_bans[1], 'blue_ban3': blue_bans[2],
        'blue_ban4': blue_bans[3], 'blue_ban5': blue_bans[4], 'red_ban1': red_bans[0], 'red_ban2': red_bans[1],
        'red_ban3': red_bans[2], 'red_ban4': red_bans[3], 'red_ban5': red_bans[4]}

blue_win = 'WIN' in soup.find('div', attrs={'class': 'col-12 blue-line-header'}).text
red_win = 'WIN' in soup.find('div', attrs={'class': 'col-12 red-line-header'}).text
games['blue_win'] = blue_win
games['red_win'] = red_win

teams = soup.find('div', attrs={'class': 'col-12 mt-4'}).find('h1').text.split(' vs ')
blue_team = teams[0]
red_team = teams[1]
tournament = soup.find('div', attrs={'class': 'col-12 col-sm-7'}).text.split('(')[0].strip()

games['blue_team'] = blue_team
games['red_team'] = red_team
games['tournament'] = tournament
df = pd.DataFrame(columns=['blue_top', 'blue_jg', 'blue_mid', 'blue_adc', 'blue_sup', 'red_top',
                          'red_jg', 'red_mid', 'red_adc', 'red_sup', 'blue_ban1', 'blue_ban2',
                          'blue_ban3', 'blue_ban4', 'blue_ban5', 'red_ban1', 'red_ban2', 'red_ban3',
                          'red_ban4', 'red_ban5', 'blue_win', 'red_win', 'blue_team', 'red_team',
                          'tournament'])
```

```python

for i in range(first, last+1):
    response = requests.get(url.format(i))
    if response.ok:
        soup = bs4.BeautifulSoup(response.content)
        champs = soup.find_all('img', attrs={'class': 'champion_icon_medium rounded-circle'})
        blue = champs[5:10]
        red = champs[15:20]
        blue = [i['alt'] for i in blue]
        red = [i['alt'] for i in red]
        blue_bans = champs[:5]
        blue_bans = [i['alt'] for i in blue_bans]
        red_bans = champs[10:15]
        red_bans = [i['alt'] for i in red_bans]
        games = {'blue_top': blue[0], 'blue_jg': blue[1], 'blue_mid': blue[2], 'blue_adc': blue[3],
             'blue_sup': blue[4], 'red_top': red[0], 'red_jg': red[1], 'red_mid': red[2], 'red_adc': red[3],
             'red_sup': red[4], 'blue_ban1': blue_bans[0], 'blue_ban2': blue_bans[1], 'blue_ban3': blue_bans[2],
            'blue_ban4': blue_bans[3], 'blue_ban5': blue_bans[4], 'red_ban1': red_bans[0], 'red_ban2': red_bans[1],
            'red_ban3': red_bans[2], 'red_ban4': red_bans[3], 'red_ban5': red_bans[4]}

        blue_win = 'WIN' in soup.find('div', attrs={'class': 'col-12 blue-line-header'}).text
        red_win = 'WIN' in soup.find('div', attrs={'class': 'col-12 red-line-header'}).text
        games['blue_win'] = blue_win
        games['red_win'] = red_win

        teams = soup.find('div', attrs={'class': 'col-12 mt-4'}).find('h1').text.split(' vs ')
        blue_team = teams[0]
        red_team = teams[1]
        tournament = soup.find('div', attrs={'class': 'col-12 col-sm-7'}).text.split('(')[0].strip()
    
        games['blue_team'] = blue_team
        games['red_team'] = red_team
        games['tournament'] = tournament
        df = df.append(games, ignore_index=True)
    else:
        continue
    
df['red_win'] = df['red_win'].astype(bool)
df['blue_win'] = df['blue_win'].astype(bool)
```
